```python
 #2
 def count_dict():
  from collections import OrderedDict
  from collections import Counter
  lis = [3, 4, 2, 1, 5, 1, 2, 4, 3, 4]
  t = Counter(lis)
  t = OrderedDict(sorted(t.items()))
  t = dict(t)
  print(t)
count_dict()
```
